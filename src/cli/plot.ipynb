```python
import json
import matplotlib.pyplot as plt
from datetime import datetime

with open('../../data/counts.json') as f:
    data = json.load(f)

counts = [item['count'] for item in data[:100]]
colors = ['red' if item['isSchilling'] else 'gray' for item in data[:100]]
dates = [datetime.fromisoformat(item['date'].replace("Z", "+00:00")) for item in data[:100]]

plt.figure(figsize=(10, 6))
plt.scatter(range(len(counts)), counts, c=colors)
plt.xlabel('')
plt.ylabel('Kommentare')
plt.title('Top 200 Artikel sortiert nach Anzahl Kommentare (rot == Schilling Bezug)')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(dates, counts, c=colors)
plt.xlabel('')
plt.ylabel('Count')
plt.title('Top 200 Artikel (nach Kommentaren) sortiert nach Datum (rot == Schilling Bezug)')
plt.xticks(rotation=45)
plt.show()
```
